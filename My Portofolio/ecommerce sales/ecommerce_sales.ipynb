Dataset yang digunakan bersumber dari ecommerce_sales.csv. Dataset tersebut mengandung data order Id,Product,Quantity Ordered, Price Each,Order Date dan Purchase Address. dataset ini yang mewakili tiap bulan transaksi penjualan.


## B. Persiapan Data dan Library yang akan digunakan

#B 1.1 Import library-library yang akan digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

#B 1.1.1 Import dataset yang akan digunakan

```python
data = pd.read_csv('sample_data/ecommerce_sales.csv')
data
```

#B 1.1.2 Menampilkan data yang mengandung Null Value(NaN) pada setiap data

```python
nan_data = data[data.isna().any(axis=1)]
print(nan_data)
```

# B 1.1.3 Hapus baris yang mengandung Null Value (NAN) pada setiap data

```python
data = data.dropna(how='all')
data
```

# B 1.1.4 Menampilkan Tipe Data 

```python
data.dtypes
```

# B 1.1.4.1 Mengubah Data dari Object Menjadi Int untuk data Quantity Ordered dan Float Price Each

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'], errors='coerce').astype('Int64')
data
```

```python
data['Price Each'] = pd.to_numeric(data['Price Each'], errors='coerce').astype('Float64')
data
```

# B 1.1.5 Mengurutkan Data dari Data Kolom Order Date

```python
data = data[data['Order Date'].str[0:2]!='Or']
data
```

# B 1.1.6 Tambahkan data month(Bulan) dengan kolom tambahan

```python
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int32')
data
```

# C. Jawaban dari setiap Pertanyaan

# Pertanyaan 1 : Pada bulan apakah perusahaan memiliki pendapatan tertinggi dan terendah? Berapa jumlah pendapatan pada bulan tersebut?

```python
data['Sales'] = data['Price Each'] * data['Quantity Ordered']
data.groupby(['Month']).sum()
```

```python
# create the data
months = range(1, 13)
print(months)

# Plot the data
plt.bar(months, data.groupby(['Month']).sum()['Sales'],color='purple')

# Add X Label on X-axis
plt.xlabel('Month')

# Add Y Label on Y-axis
plt.ylabel('Sales in USD($)')

# Add a title to graph
plt.title("Sales Rating Distribution")
plt.xticks(months)

# Show the plot
plt.show()
```

# Pertanyaan 2 : Produk apakah yang laris terjual dan tidak laris terjual? 

```python
# create the data
product_group = data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
keys = [pair for pair, df in product_group]

# Plot the data
plt.bar(keys, quantity_ordered,color='purple')

# Add X Label on X-axis
plt.xlabel('Product')

# Add Y Label on Y-axis
plt.ylabel('Quantity Ordered')

# Add a title to graph
plt.title("Best Seller Products")
plt.xticks(keys, rotation='vertical', size=8)

# Show the plot
plt.show()
```

# Pertanyaan 3 : kapan waktu yang cocok bagi tim marketing untuk menampilkan iklan?

```python
data['Hour'] = pd.to_datetime(data['Order Date']).dt.hour
data['Minute'] = pd.to_datetime(data['Order Date']).dt.minute
data['Count'] = 1
data
```

```python
# create the data
keys = [pair for pair, data in data.groupby(['Hour'])]

# Plot the data
plt.plot(keys, data.groupby(['Hour']).count()['Count'],color='purple')

#Add X Label on X-axis
plt.xlabel('Hour')

# Add Y Label on Y-axis
plt.ylabel('Count')

# Add a title to graph
plt.title("Time to Sell")
plt.xticks(keys)
plt.grid()

# Show the plot
plt.show()
```
